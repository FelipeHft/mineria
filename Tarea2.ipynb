# Ejercicio 1

Escribir un programa que solicite un número y calcule el cuadrado de ese número.
- Si el cuadrado es 100 o mayor, imprima el valor al cuadrado y la palabra 'grande'.
- Si el cuadrado está entre 50 y 99, imprima el valor al cuadrado y la palabra 'medio'.
- De lo contrario, simplemente imprima 'demasiado pequeño para molestarlo'.

```python
x = input("Ingrese un numero: ")
cuadrado = pow(x, 2)
if(cuadrado >= 100):
    print(cuadrado, "grande") 
elif((cuadrado >= 50) and (cuadrado <= 99)):
    print(cuadrado, "medio")
else:
    print("demasiado pequeño para molestarlo")
```

# Ejercicio 2

A continuación se muestra el código que crea una lista de cuatro elementos, representados como diccionarios que especifican el color y tamaño.

items = [{'color':'red', 'size':10}, {'color':'blue', 'size':3},{'color':'green', 'size':15}, {'color':'yellow', 'size':6}]


Añadir código para pedir un número, luego imprimir el color de todos elementos cuyo tamaño es mayor que ese número.

```python
items = [{'color':'red', 'size':10}, {'color':'blue', 'size':3},{'color':'green', 'size':15}, {'color':'yellow', 'size':6}]
x = input("Ingrese un numero: ")
for i in items:
    if(i['size'] > x):
        print(i['color'])
```

# Ejercicio 3

Crear un bucle que solicite dos números, e imprima los colores de todos los elementos cuyo tamaño está entre los dos números, es decir, cuyo tamaño es mayor que el primer número e inferior al segundo número.

Si el segundo número del usuario es igual o menor que el primer número, el programa debe imprimir 'Hecho' y detenerse.

```python
items = [{'color':'red', 'size':10}, {'color':'blue', 'size':3},{'color':'green', 'size':15}, {'color':'yellow', 'size':6}]
fin = 0
while(fin == 0):
    x = input("Ingrese numero 1: ")
    y = input("Ingrese numero 2: ")
    if(y <= x):
        print("Hecho")
        fin=1
    else:
        for i in items:
            if((i['size'] > x) and (i['size'] < y)):
                print(i['color'])
```

#### 3.- ¿Cuál es el promedio, máximo y mínimo de pases de los defensas y delanteros?

```python
print "minimo de pases de los defensas: ", min(players[players.position == "defender"]['passes'])
print "maximo de pases de los defensas: ", max(players[players.position == "defender"]['passes'])
print "promedio de pases de los defensas: ", players[players.position == "defender"]['passes'].mean()
```

```python
print "minimo de pases de los delanteros: ", min(players[players.position == "forward"]['passes'])
print "maximo de pases de los delanteros: ", max(players[players.position == "forward"]['passes'])
print "promedio de pases de los delanteros: ", players[players.position == "forward"]['passes'].mean()
```

#### 4.- ¿Qué equipo tiene el mayor ratio de 'goalsFor' y 'goalsAgainst'? Imprimir sólo el nombre del equipo.

```python
ratioMayor = 0
ind = 0
for index, row in teams.iterrows():
    ratio = row.goalsFor / row.goalsAgainst
    if(ratio > ratioMayor):
        ratioMayor = ratio
        ind = index

print("El equipo con el ratio mayor es: ", teams['team'].iloc[ind])

```

#### 5.- ¿Cuántos jugadores que juegan en un equipo con ranking <5 jugaron más de 400 minutos?


```python
playersTeams = pd.merge(players, teams, on = 'team')
playersTeams[(playersTeams.ranking < 5) & (playersTeams.minutes > 400)]['surname'].count()
```

#### 6.- Calcule por cada equipo la diferencia entre el jugador que da menos pases con el que da más pases. (use función groupby, agg con función lambda)

```python
groupPT = playersTeams.groupby('team')
groupPT['passes'].agg(lambda x: max(x) - min(x))

```
